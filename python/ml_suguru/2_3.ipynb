```python
from sklearn.model_selection import train_test_split
from mglearn.datasets import load_extended_boston
```

```python
X,y = load_extended_boston()

X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0)
```

```python
import matplotlib.pyplot as plt
from sklearn import linear_model
plt.style.use("seaborn-darkgrid")
```

```python

X_train_single=X_train[:,5].reshape(-1,1)
X_test_single = X_test[:,5].reshape(-1,1)
```

```python
lm_single = linear_model.LinearRegression()
lm_single.fit(X_train_single,y_train)

y_pred_train = lm_single.predict(X_train_single)
```

```python
# 訓練 データ と 学習 し た 単 回帰 モデル の 可視化 
plt.xlabel(' RM') 
# x 軸 の ラベル を 設定 
plt.ylabel(' MEDV') 
# y 軸 の ラベル を 設定 
plt.scatter( X_train_single, y_train) 
# 各 データ 点（ 散布 図） を 作図 
plt.plot( X_train_single, y_pred_train, color ='red', linewidth = 2) 
# 回帰 し た 直線 を 作図 
plt.show() # 作成 し た グラフ を 描画
```

```python
print("intercept:{}".format(lm_single.intercept_))
print("Train score:{}".format(lm_single.score(X_train_single,y_train)))
print("Test score:{}".format(lm_single.score(X_test_single,y_test)))
```
